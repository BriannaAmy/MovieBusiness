```python
import requests
import json
from config import api_key
from pprint import pprint
import pandas as pd
import csv
import time
```

```python
# Configuration of API variables
url = "http://www.omdbapi.com/?t="
api_key = "&apikey=" + api_key

# Query each movie title and populate variables
with open("Resources/MovieList.csv", "r") as file:
    
    movies = file.readlines()
    
    #variable lists creation
    movieTitle = []
    genre = []
    runtime = []
    boxOffice = []
    metaRating = []
    imdbRating = []
    releaseYear = []
    
    for movie in movies:
    
        movie = movie.replace(" ","+")
        queryURL = url + movie + api_key
        
        try:
            
            response = requests.get(queryURL).json()
            
            # create variables for Data Frame
            movieTitle.append(response.get('Title', 'N/A'))
            genre.append(response.get('Genre', 'N/A'))
            runtime.append(response.get('Runtime', 'N/A'))
            boxOffice.append(response.get('BoxOffice', 'N/A'))
            metaRating.append(response.get('Metascore', 'N/A'))
            imdbRating.append(response.get('imdbRating', 'N/A'))
            releaseYear.append(response.get('Year', 'N/A'))
            
        except:
            print(f"{movie} not found...skipping")
            pass
        
        time.sleep(2)
        
#populate data frame
movieDataDF = pd.DataFrame({
    "Movie Title": movieTitle,
    "Genre": genre,
    "Runtime (min)": runtime,
    "Box Office ($)": boxOffice,
    "Metacritic Rating": metaRating,
    "IMDb Rating": imdbRating,
    "Release Year": releaseYear
})

movieDataDF.head()
```
